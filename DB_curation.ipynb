```python
import pandas as pd
data = pd.read_csv("220629_dupl_rmvd.csv")
data = data[~data.No.isin([48, 199, 1155, 517, 518, 931])]
data = data[data["Mole/Weight/Volume Fraction"] == "mole"]
data = data.iloc[:-15]
data = data[~data.blends_str_repr.str.contains('Other')]
```

```python
data
```

```python
for i in range(1,14):
    data['smi_mol'+str(i)] = data['smi_mol'+str(i)].fillna('[HH]')
    data['can_smi_'+str(i)] = data['can_smi_'+str(i)].fillna('[HH]')
    data['x_mol_'+str(i)] = data['x_mol_'+str(i)].fillna('0.0')
```
